```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, KFold
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
from scipy.stats import norm
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.model_selection import cross_val_score, train_test_split
import matplotlib.pyplot as plt

class BayesianOptimizer:
    def __init__(self, param_bounds, n_iter=50, init_samples=5):
        self.param_bounds = param_bounds
        self.n_iter = n_iter
        self.init_samples = init_samples
        self.kernel = RBF()
        self.gpr = GaussianProcessRegressor(kernel=self.kernel, alpha=1e-6, normalize_y=True, n_restarts_optimizer=10)
        self.samples = []
        self.scores = []

    def sample_params(self):
        return {key: np.random.uniform(low, high) for key, (low, high) in self.param_bounds.items()}

    def fit_gpr(self):
        X = np.array([list(params.values()) for params in self.samples])
        y = np.array(self.scores)
        self.gpr.fit(X, y)

    def expected_improvement(self, x):
        mu, sigma = self.gpr.predict([x], return_std=True)
        mu_sample = self.gpr.predict(np.array([list(params.values()) for params in self.samples]))
        sigma = sigma.reshape(-1, 1)
        mu_sample_opt = np.max(mu_sample)

        with np.errstate(divide='warn'):
            imp = mu - mu_sample_opt - 0.01
            Z = imp / sigma
            ei = imp * norm.cdf(Z) + sigma * norm.pdf(Z)
            ei[sigma == 0.0] = 0.0
        return ei

    def propose_location(self):
        best_ei = 0
        best_x = None
        for _ in range(100):
            x = np.array([np.random.uniform(low, high) for low, high in self.param_bounds.values()])
            ei = self.expected_improvement(x)
            if ei > best_ei:
                best_ei = ei
                best_x = x
        return best_x

    def optimize(self, objective_function):
        for _ in range(self.init_samples):
            sample = self.sample_params()
            score = objective_function(sample)
            self.samples.append(sample)
            self.scores.append(score)

        self.fit_gpr()

        for _ in range(self.n_iter):
            x = self.propose_location()
            sample = {key: x[i] for i, key in enumerate(self.param_bounds.keys())}
            score = objective_function(sample)
            self.samples.append(sample)
            self.scores.append(score)
            self.fit_gpr()

        best_sample = self.samples[np.argmax(self.scores)]
        return best_sample
```

```python
import numpy as np
from hyperopt import fmin, tpe, hp, Trials, STATUS_OK
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

class BayesianOptimizerHyperopt:
    def __init__(self, param_bounds, n_iter=10, init_samples=5):
        self.param_bounds = param_bounds
        self.n_iter = n_iter
        self.init_samples = init_samples
        self.samples = []
        self.scores = []
        self.trials = Trials()

    def sample_params(self):
        return {key: np.random.uniform(low, high) for key, (low, high) in self.param_bounds.items()}

    def optimize(self, objective_function):
        space = {key: hp.uniform(key, low, high) for key, (low, high) in self.param_bounds.items()}

        def objective(params):
            score = objective_function(params)
            return {'loss': -score, 'status': STATUS_OK}

        best = fmin(objective,
                    space=space,
                    algo=tpe.suggest,
                    max_evals=self.n_iter,
                    trials=self.trials)

        return best

def objective_function(params):
    rf = RandomForestClassifier(
        n_estimators=int(params['n_estimators']),
        max_depth=int(params['max_depth']),
        min_samples_split=int(params['min_samples_split']),
        min_samples_leaf=int(params['min_samples_leaf']),
        random_state=42
    )
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(rf, X, y, cv=cv, scoring='accuracy')
    return np.mean(scores)

# Define the parameter bounds
param_bounds = {
    'n_estimators': (10, 200),
    'max_depth': (1, 20),
    'min_samples_split': (2, 20),
    'min_samples_leaf': (1, 20)
}

# Initialize the Bayesian Optimizer with hyperopt
optimizer_hyperopt = BayesianOptimizerHyperopt(param_bounds, n_iter=10, init_samples=5)

# Optimize the parameters
best_params_hyperopt = optimizer_hyperopt.optimize(objective_function)

print("Best Parameters found by Hyperopt:")
print(best_params_hyperopt)

# Train the final model with the best parameters
rf = RandomForestClassifier(
    n_estimators=int(best_params_hyperopt['n_estimators']),
    max_depth=int(best_params_hyperopt['max_depth']),
    min_samples_split=int(best_params_hyperopt['min_samples_split']),
    min_samples_leaf=int(best_params_hyperopt['min_samples_leaf']),
    random_state=42
)

rf.fit(X, y)
score_hyperopt = rf.score(X, y)
print("Final Model Accuracy with Hyperopt:", score_hyperopt)
```

```python

import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from bayes_opt import BayesianOptimization
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

features = pd.read_csv('application_train.csv')

features = features.sample(n=16000, random_state=42)

features = features.select_dtypes('number')

labels = np.array(features['TARGET'].astype(np.int32))
features = features.drop(columns=['TARGET', 'SK_ID_CURR'])

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=6000, random_state=42)

print('Train shape: ', train_features.shape)
print('Test shape: ', test_features.shape)

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)

def evaluate_model(model, X, y, cv_splits=5):
    cv = KFold(n_splits=cv_splits, shuffle=True, random_state=42)
    scores = cross_val_score(model, X, y, cv=cv, scoring='roc_auc')
    return scores

baseline_model = lgb.LGBMClassifier(random_state=42)
baseline_scores = evaluate_model(baseline_model, train_features, train_labels)
baseline_model.fit(train_features, train_labels)
baseline_preds = baseline_model.predict_proba(test_features)[:, 1]
baseline_auc = roc_auc_score(test_labels, baseline_preds)

print("Baseline Model Cross-validation ROC AUC scores:", baseline_scores)
print("Baseline Model Mean ROC AUC:", baseline_scores.mean())
print("Baseline Model Test ROC AUC:", baseline_auc)
```

```python
def bayesian_opt_lgb(train_features, train_labels):
    def objective(n_estimators, max_depth, learning_rate, num_leaves, feature_fraction, bagging_fraction):
        params = {
            'boosting_type': 'gbdt',
            'objective': 'binary',
            'n_estimators': int(n_estimators),
            'max_depth': int(max_depth),
            'learning_rate': learning_rate,
            'num_leaves': int(num_leaves),
            'feature_fraction': feature_fraction,
            'bagging_fraction': bagging_fraction,
            'random_state': 42,
            'verbosity': -1
        }

        cv = KFold(n_splits=5, shuffle=True, random_state=42)
        scores = []

        for train_idx, val_idx in cv.split(train_features, train_labels):
            X_train, X_val = train_features[train_idx], train_features[val_idx]
            y_train, y_val = train_labels[train_idx], train_labels[val_idx]

            model = lgb.LGBMClassifier(**params)
            model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric='auc')

            val_preds = model.predict_proba(X_val)[:, 1]
            score = roc_auc_score(y_val, val_preds)
            scores.append(score)

        return np.mean(scores)

    optimizer = BayesianOptimization(
        f=objective,
        pbounds={
            'n_estimators': (10, 1000),
            'max_depth': (1, 20),
            'learning_rate': (0.01, 0.3),
            'num_leaves': (20, 100),
            'feature_fraction': (0.1, 0.9),
            'bagging_fraction': (0.8, 1.0)
        },
        random_state=42
    )

    optimizer.maximize(init_points=10, n_iter=10)
    best_params = optimizer.max['params']
    best_params['n_estimators'] = int(best_params['n_estimators'])
    best_params['max_depth'] = int(best_params['max_depth'])
    best_params['num_leaves'] = int(best_params['num_leaves'])

    return best_params

# Run Bayesian Optimization
bayesian_best_params = bayesian_opt_lgb(train_features, train_labels)
bayesian_model = lgb.LGBMClassifier(**bayesian_best_params, random_state=42)
bayesian_scores = evaluate_model(bayesian_model, train_features, train_labels)
bayesian_model.fit(train_features, train_labels)
bayesian_preds = bayesian_model.predict_proba(test_features)[:, 1]
bayesian_auc = roc_auc_score(test_labels, bayesian_preds)

print("Bayesian Optimizer Cross-validation ROC AUC scores:", bayesian_scores)
print("Bayesian Optimizer Mean ROC AUC:", bayesian_scores.mean())
print("Bayesian Optimizer Test ROC AUC:", bayesian_auc)
```

```python

import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from hyperopt import hp, fmin, tpe, Trials, STATUS_OK
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

features = pd.read_csv('/content/application_train.csv')

features = features.sample(n=16000, random_state=42)

features = features.select_dtypes('number')

labels = np.array(features['TARGET'].astype(np.int32))
features = features.drop(columns=['TARGET', 'SK_ID_CURR'])

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=6000, random_state=42)

print('Train shape: ', train_features.shape)
print('Test shape: ', test_features.shape)

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

```python

def hyperopt_lgb(train_features, train_labels):
    def objective(params):
        params = {
            'boosting_type': 'gbdt',
            'objective': 'binary',
            'n_estimators': int(params['n_estimators']),
            'max_depth': int(params['max_depth']),
            'learning_rate': params['learning_rate'],
            'num_leaves': int(params['num_leaves']),
            'feature_fraction': params['feature_fraction'],
            'bagging_fraction': params['bagging_fraction'],
            'random_state': 42,
            'verbosity': -1
        }

        cv = KFold(n_splits=5, shuffle=True, random_state=42)
        scores = []

        for train_idx, val_idx in cv.split(train_features, train_labels):
            X_train, X_val = train_features[train_idx], train_features[val_idx]
            y_train, y_val = train_labels[train_idx], train_labels[val_idx]

            model = lgb.LGBMClassifier(**params)
            model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric='auc')

            val_preds = model.predict_proba(X_val)[:, 1]
            score = roc_auc_score(y_val, val_preds)
            scores.append(score)

        return np.mean(scores)
    space = {
        'n_estimators': hp.quniform('n_estimators', 10, 1000, 1),
        'max_depth': hp.quniform('max_depth', 1, 20, 1),
        'learning_rate': hp.uniform('learning_rate', 0.01, 0.3),
        'num_leaves': hp.quniform('num_leaves', 20, 100, 1),
        'feature_fraction': hp.uniform('feature_fraction', 0.1, 0.9),
        'bagging_fraction': hp.uniform('bagging_fraction', 0.8, 1.0)
    }

    trials = Trials()
    best = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=50, trials=trials, rstate=np.random.default_rng(42))

    best['n_estimators'] = int(best['n_estimators'])
    best['max_depth'] = int(best['max_depth'])
    best['num_leaves'] = int(best['num_leaves'])

    return best

hyperopt_best_params = hyperopt_lgb(train_features, train_labels)
hyperopt_model = lgb.LGBMClassifier(**hyperopt_best_params, random_state=42)
hyperopt_model.fit(train_features, train_labels)
hyperopt_scores = cross_val_score(hyperopt_model, train_features, train_labels, cv=5, scoring='roc_auc')

print("Hyperopt Optimizer Cross-validation ROC AUC scores:", hyperopt_scores)
print("Hyperopt Optimizer Mean ROC AUC:", hyperopt_scores.mean())

hyperopt_preds = hyperopt_model.predict_proba(test_features)[:, 1]
hyperopt_auc = roc_auc_score(test_labels, hyperopt_preds)
print("Hyperopt Optimizer Test ROC AUC:", hyperopt_auc)
```
